```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
database= ('../DATA/database.sqlite')
```

## Make connection

```python
conn= sqlite3.connect(database)
```

```python
conn
```

## Look at the tables

```python
tables=pd.read_sql('''SELECT *
                        FROM sqlite_master
                        WHERE type="table"''',conn)
tables
```

### SELECT A COUNTRY

```python
country=pd.read_sql('''SELECT *
                FROM country''',conn)
country
```

### SELECT LatestIndustrialData	greater than 2011

```python
pd.read_sql('''SELECT Shortname AS country, LatestIndustrialData
                FROM country
                WHERE 'LatestIndustrialData' >"2011"''',conn)
                
```

### SELECT NEWZEALAND, REGIN, industrial data greater than 2009,latesttradedata

```python
pd.read_sql('''SELECT Shortname AS country,
                    region,
                    latesttradedata,
                    latestindustrialdata
                FROM country
                WHERE shortname= "New Zealand" 
                AND latesttradedata>"2003" 
                AND latestindustrialdata>"2009"''',conn)
                    
```

```python
pd.read_sql('''SELECT Shortname AS country,
                    region,
                    latesttradedata,
                    latestindustrialdata
                FROM country
                WHERE shortname= "New Zealand" 
                OR latesttradedata>"2013" 
                OR latestindustrialdata>"2009"''',conn)
```

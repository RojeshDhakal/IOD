```python
df['Age'].isnull().sum()
```

```python
#Impute null value
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Age'].isnull().sum()
```

```python
df['Sex'].fillna(df['Sex'].mode(),inplace=True)
df['Sex'].isnull().sum()
```

```python
#modify/add new columns (Changing the value of male and female from sex column)
df['Sex']=df['Sex'].map({"male":'0',"female":"1"})
df.head()
```

```python
#accesssing the first and last name from name column
df['last_name']= df['Name'].apply(lambda x: x.split(',')[0])
df['first_name']= df['Name'].apply(lambda x:''.join (x.split (',')[1]))
```

```python
df.head()
```

```python
#Sets to 1 for men in 3rd class
df['Thrid&Men']=df.apply(lambda row: int(row['Pclass']==3 and row['Sex']=="0"),axis=1)
df.head()
```

```python
def findAgeGroup(age):
    if age<18:
        return 1
    elif age>=18 and age<40:
        return 2
    elif age>=40 and age<60:
        return 3
    else:
        return 4
df['Age_group']=df['Age'].apply(lambda x: findAgeGroup(x))
df.head()
```

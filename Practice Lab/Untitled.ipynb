```python
import pandas as pd
import numpy as np
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
data = pd.read_csv("eda.txt", names=cols)
```

```python
data['normalized-losses']=data['normalized-losses'].replace('?', np.NaN)
```

```python
data.isnull().sum()
```

```python
data=data.replace('?', np.NaN)
```

```python
data.isnull().sum()
```

```python
data['normalized-losses']= data['normalized-losses'].astype(float)
```

```python
data['normalized-losses'].describe().T
```

```python
data['normalized-losses']= data['normalized-losses'].replace(np.NaN, data['normalized-losses'].mean())
```

```python
data.isnull().sum()
```

```python
data['bore']= data['bore'].astype('float')
```

```python
data.bore.info()
```

```python
data['bore'].replace(np.NaN, data['bore'].mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
##changing multiple columns at one time

data[['stroke','horsepower']]= data[['stroke','horsepower']].astype('float')
```

```python
data.info()
```

```python
data['stroke']= data['stroke'].replace(np.NaN,data['stroke'].mean())
```

```python
data.stroke.isnull().sum()
```

```python
data['horsepower']= data['horsepower'].replace(np.NaN,data['horsepower'].mean())
```

```python
data.horsepower.isnull().sum()
```

```python
data[['bore','stroke' ,'compression-ratio','horsepower']].corr().T
```

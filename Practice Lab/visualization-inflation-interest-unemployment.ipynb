**Creating a new Dataframe with continuous data**

```python
continuous= df[['Inflation, consumer prices (annual %)',
       'Inflation, GDP deflator (annual %)', 'Real interest rate (%)',
       'Deposit interest rate (%)', 'Lending interest rate (%)',
       'Unemployment, total (% of total labor force) (national estimate)',
       'Unemployment, total (% of total labor force) (modeled ILO estimate)']]
```

```python
continuous.describe()
```

**Removing Null Values**

```python
continuous_rm = continuous.dropna()
```

**Histogram Plot**

```python
continuous_rm.hist(figsize=(15,15))
```

**Box Plot**

```python
continuous_rm.plot.box(subplots = True,layout=(4,3), figsize=(10,15))
```

**Removing Outliers**

```python
z = np.abs(stats.zscore(continuous_rm))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

```python
continuous_clean = continuous_rm[(z < 3).all(axis=1)]
```

**Comparing histogram before and after removing outliers**

```python
continuous_rm.plot.box(subplots = True,layout=(4,3), figsize=(10,15))
continuous_clean.plot.box(subplots = True,layout=(4,3), figsize=(10,15))
```

**Correlation Plot**

```python
plt.figure(figsize=(10,10))
sns.heatmap(continuous_clean.corr(),annot=True,cmap="Blues")
```

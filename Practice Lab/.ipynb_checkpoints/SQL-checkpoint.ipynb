```python
#CHecking for missing value
pd.read_sql(""" SELECT COUNT(*)
                FROM Country
                WHERE LATESTTRADEDATA IS  NULL ;""",
           conn)
```

```python
pd.read_sql("""SELECT COUNT(*) - COUNT(CountryCode) AS missing_Country_code,
                      COUNT(*) - COUNT(CurrencyUnit) AS missing_Currency,
                      COUNT(*) - COUNT(Region) AS missing_TableName,
                FROM Country;""",
            conn)
```

```python
#how many distinct currency 
pd.read_sql(""" SELECT COUNT(DISTINCT CurrencyUnit) AS Currency_Worldwide
                FROM Country ;""",
           conn)
```

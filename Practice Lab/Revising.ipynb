```python
import numpy as np
np.random.normal(3,2,5)
```

random.normal gives the random number. the first number is mean and second is std and 3rd number is the size. rows and colums can be generated by specifying in paranathesis

```python
np.random.normal(3,2,(4,4))
```

```python
np.round(np.random.normal(3,2,5))
```
